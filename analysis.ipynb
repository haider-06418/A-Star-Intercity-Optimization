# Analysis

```python
import json
import time
import matplotlib.pyplot as plt
import seaborn as sns

import a_star
import dijkstra
import bestfs
```

```python
# function to load json file into a dictionary
def load_json(json_file):
    with open(json_file, 'r') as file:
        data = json.load(file)
    return data


# plotting a bar chart
def plot_bar_chart(categories, values, title, ylabel):
    sns.set(style="whitegrid")
    # plt.figure(figsize=(10, 8))
    sns.barplot(x=categories, y=values, palette="viridis")
    plt.title(title)
    plt.ylabel(ylabel)
    plt.show()
```

```python
graph_file = "graphs/graph_updated.json"
graph = load_json(graph_file)
```

```python
# measuring performance metrics of for all algorithms for the given start and end nodes

def main(graph, start_city, goal_city):

    algorithms = ['A*', 'Dijkstra', 'BestFS']
    times = []
    path_costs = []
    nodes_explored = []
    paths = []

    # a star
    start_time = time.time()
    path_1, cost_1, no_of_nodes_explored  = a_star.a_star_search(graph, start_city, goal_city) 
    end_time = time.time()
    time_taken = end_time - start_time
    times.append(time_taken)
    path_costs.append(cost_1)
    nodes_explored.append(no_of_nodes_explored)
    paths.append(path_1)
    print('A*:')
    if path_1:
        print(f"Path from {start_city} to {goal_city}:")
        print(" -> ".join(path_1))
        print(f"Total Cost: {cost_1}")
        print(f"Number of nodes explored: {no_of_nodes_explored}")
        print(f"Time taken: {time_taken} seconds")
    else:
        print(f"No path found from {start_city} to {goal_city}.")
    print(' ')


    # djkistra 
    start_time = time.time()
    path_2, cost_2, no_of_nodes_explored = dijkstra.dijkstra_algorithm(graph, start_city, goal_city)
    end_time = time.time() 
    time_taken = end_time - start_time
    times.append(time_taken)
    path_costs.append(cost_2)
    nodes_explored.append(no_of_nodes_explored)
    paths.append(path_2)
    print('Dijkstra:')
    if path_2:
        print(f"Path from {start_city} to {goal_city}:")
        print(" -> ".join(path_2))
        print(f"Total Cost: {cost_2}")
        print(f"Number of nodes explored: {no_of_nodes_explored}")
        print(f"Time taken: {time_taken} seconds")
    else:
        print(f"No path found from {start_city} to {goal_city}.")
    print(' ')



    # best first search
    start_time = time.time()
    path_3, cost_3, no_of_nodes_explored = bestfs.best_first_search(graph, start_city, goal_city)
    end_time = time.time()
    time_taken = end_time - start_time
    times.append(time_taken)
    path_costs.append(cost_3)
    nodes_explored.append(no_of_nodes_explored)
    paths.append(path_3)
    print('BestFS:')
    if path_3:
        print(f"Path from {start_city} to {goal_city}:")
        print(" -> ".join(path_3))
        print(f"Total Cost: {cost_3}")
        print(f"Number of nodes explored: {no_of_nodes_explored}")
        print(f"Time taken: {time_taken} seconds")
    else:
        print(f"No path found from {start_city} to {goal_city}.")
    print(' ')


    # plotting each metric
    plot_bar_chart(algorithms, nodes_explored, 'Nodes Explored Comparison', 'Number of Nodes')
    plot_bar_chart(algorithms, path_costs, 'Path Cost Comparison', 'Cost')
    plot_bar_chart(algorithms, times, 'Execution Time Comparison', 'Time (seconds)')
```

```python
# results category: bestfs highest cost

start_city = "Loralai"
goal_city = "Jamshoro"

main(graph, start_city, goal_city)

# start_city = "Sanghar"
# goal_city = "Khuzdar"

# main(graph, start_city, goal_city)
```
